Generator: transformer output (batch_size, seq_len, d_model) -> Linear(d_model, vocab_size) -> softmax(dim=-1)

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Generator(nn.Module):
    def __init__(self, d_model, vocab):
        super(Generator, self).__init__()
        self.proj = nn.Linear(in_features=d_model, out_features=vocab)
    
    def forward(self, x):
        print(f"+GENERATOR:x.shape={x.shape}")
        proj_x = self.proj(x)
        g_out = F.log_softmax(proj_x, dim=-1)
        print(f"-GENERATOR:g_out.shape={g_out.shape}")
        return g_out
```

```python
batch_size, seq_len, d_model, vocab = 2, 10, 512, 10

x = torch.randn(batch_size, seq_len, d_model)
generator = Generator(d_model, vocab)
y = generator(x)
# get the last token logits
logits = y[:, -1]
print(f"Logits shape: {logits.shape}\n{logits}")
next_token = torch.argmax(logits, dim=-1)
print(f"Next token shape: {next_token.shape} \n{next_token}")
```
