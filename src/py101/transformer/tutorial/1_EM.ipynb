Embeddings
is a lookup table that maps discrete tokens(like word IDs) to continuous vector representations.

```python
import torch
import torch.nn as nn
import math
```

```python
class Embeddings(nn.Module):
    def __init__(self, vocab, d_model):
        super(Embeddings, self).__init__()
        # learnable lookup table shape (vocab, d_model)
        self.embedding = nn.Embedding(vocab, d_model)
        self.d_model = d_model

    def forward(self, x):
        # x shape (sequences, seq_len)
        # embeddings shape (sequences, seq_len, d_model)
        return self.embedding(x) * math.sqrt(self.d_model)
```

```python
# 10 unique token vocabulary
vocab = 10
# each token represented by 4-dimensional vector
d_model = 4

embedding = nn.Embedding(vocab, d_model)

print(f"Embedding weights shape: {embedding.weight.shape}")
print(f"Embedding weights: {embedding.weight}")
```

```python
# Single Token
token = torch.tensor([3])

# Forward pass: lookup the embedding
token_embed = embedding(token)
print(f"input token shape: {token.shape}")
print(f"embedding shape: {token_embed.shape}")
print(f"embedding vector: {token_embed}")
```

```python
# Input: Sentence with tokens
# "I love machine learning" -> [2, 5, 1, 8]
tokens = torch.tensor([2, 5, 1, 8])

# Forward pass: lookup the embedding
tokens_embed = embedding(tokens)
print(f"input tokens shape: {tokens.shape}")
print(f"embeddings shape: {tokens_embed.shape}")
print(f"embeddings vector: {tokens_embed}")

# each token's embedding vector
for i, token_id in enumerate(tokens):
    print(f"Token {token_id}: {tokens_embed[i]}")
```

```python
batch_tokens = torch.tensor([[[2, 5, 1, 8, 3], # sentence 1
                             [7, 2, 9, 1, 4]]] # sentence 2
                            )
# Forward pass: lookup the embedding
batch_tokens_embed = embedding(batch_tokens)
print(f"input tokens shape: {batch_tokens.shape}")
print(f"embeddings shape: {batch_tokens_embed.shape}")
print(f"embeddings vector: {batch_tokens_embed}")
```
